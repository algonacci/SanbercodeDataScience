Jelaskan perbedaan antara Univariate, Bivariate dan Multivariate Analysis 

Jawaban:

- Univariate Analysis, adalah analisis yang dilakukan untuk satu variabel atau per variabel.
- Bivariate Analysis, adalah analisis yang dilakukan untuk menganalisis hubungan dua variabel.
- Multivariate Analysis, adalah analisis yang dilakukan untuk menganalisis hubungan lebih dari dua variabel.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df
```

```python
df.info()
```

# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()



```python
#code here
df.dropna()
df_mean = df.groupby(['species']).mean()
df_mean
```
